```python
from detoxify import Detoxify
import pandas as pd
from tqdm import tqdm
```

```python
df1 = pd.read_csv('/Users/hgardner/Desktop/toxicity/tox_t.csv')
df2 = pd.read_csv('/Users/hgardner/Desktop/toxicity/data/preprocessed_data/cat_title.csv')
```

```python
df = df1.merge(df2, on='bibid', how='left')
```

```python
df = df.drop(['title_y','Unnamed: 0'], axis=1)
```

```python
model = Detoxify('multilingual',device='cpu')
tqdm.pandas()
```

```python
df['identity_attack'] = df['clean_title'].progress_map(lambda line: model.predict(line)['identity_attack'])
```
